```python
#@title import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#@title Read the data
df=pd.read_excel("/content/Raw Data.xlsx")
df
```

```python
#@title display the first 5 rows of the data
df.head()
```

```python
#@title Show basic insights of the Data
df.info()
```

```python
#@title show basic statistical insights about the data
df.describe()
```

```python
#@title check missing values
df.isnull().sum()
```

```python
#@title Check duplicated
df.duplicated().sum()
```

```python
#@title Remove duplicates and Check again
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
#@title Exploratory Data Analysis
```

```python
#@title Countplots for Categorical Columns:

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.countplot(x='Marital Status', data=df)

plt.subplot(2, 2, 2)
sns.countplot(x='Gender', data=df)

plt.subplot(2, 2, 3)
sns.countplot(x='Education', data=df)
plt.xticks(rotation=45, ha="right")

plt.subplot(2, 2, 4)
sns.countplot(x='Occupation', data=df)
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
#@title Income Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Income'], bins=20, kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
#@title Home Ownership Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Home Owner', data=df)
plt.title('Home Ownership Distribution')
plt.show()
```

```python
#@title Car ownership Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Cars', data=df)
plt.title('Car Ownership Distribution')
plt.xlabel('Number of Cars')
plt.ylabel('Count')
plt.show()
```

```python
#@title Correlation Heatmap
import seaborn as sns
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
